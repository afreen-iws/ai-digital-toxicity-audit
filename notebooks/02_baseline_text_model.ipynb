```python
from pathlib import Path
import json
import pandas as pd

train_file = Path("data/raw/hateful_memes/data/train.jsonl")
dev_file   = Path("data/raw/hateful_memes/data/dev.jsonl")

print("Train file exists:", train_file.exists())
print("Dev file exists:  ", dev_file.exists())

def load_jsonl(path):
    with open(path, "r", encoding="utf-8") as f:
        return [json.loads(line) for line in f if line.strip()]

df_train = pd.DataFrame(load_jsonl(train_file))
df_val   = pd.DataFrame(load_jsonl(dev_file))

df_train.head(), df_val.head()
```

```python
!ls -R data/raw/hateful_memes
```

```python
# Keep only the columns we need for text baseline
df_train = df_train[["text", "label"]].dropna()
df_val   = df_val[["text", "label"]].dropna()

print(df_train.shape, df_val.shape)
df_train.head()
```

```python
from datasets import Dataset

train_ds = Dataset.from_pandas(df_train.reset_index(drop=True))
val_ds   = Dataset.from_pandas(df_val.reset_index(drop=True))

train_ds, val_ds
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
```

```python
def tokenize_batch(batch):
    return tokenizer(
        batch["text"],
        padding="max_length",
        truncation=True,
        max_length=128
    )

train_ds = train_ds.map(tokenize_batch, batched=True)
val_ds   = val_ds.map(tokenize_batch, batched=True)
```

```python
# Hugging Face will add an internal index column; we don’t need it for training
train_ds = train_ds.remove_columns(["__index_level_0__"]) if "__index_level_0__" in train_ds.column_names else train_ds
val_ds   = val_ds.remove_columns(["__index_level_0__"]) if "__index_level_0__" in val_ds.column_names else val_ds

# Tell the dataset which columns to return as tensors
train_ds.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "label"]
)
val_ds.set_format(
    type="torch",
    columns=["input_ids", "attention_mask", "label"]
)

train_ds[0]
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=2
)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    acc = accuracy_score(labels, preds)
    f1 = f1_score(labels, preds)
    return {"accuracy": acc, "f1": f1}
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./models/checkpoints/bert_text_baseline",
    eval_strategy="epoch",        # evaluate at end of each epoch
    save_strategy="epoch",             # save checkpoint each epoch
    num_train_epochs=2,                # you can increase to 3–4 later
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    learning_rate=2e-5,
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    greater_is_better=True,
    report_to="none"                   # we’re not using wandb/tensorboard
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    compute_metrics=compute_metrics,
)

trainer.train()
```

```python
eval_results = trainer.evaluate()
eval_results
```

```python
# Get raw predictions on validation set
pred_output = trainer.predict(val_ds)
logits = pred_output.predictions
y_pred = np.argmax(logits, axis=-1)
y_true = np.array(df_val["label"])

print("Classification report:\n")
print(classification_report(y_true, y_pred, digits=4))

print("Confusion matrix:\n")
print(confusion_matrix(y_true, y_pred))
```

```python
# Attach predictions to the original validation DataFrame
df_val_pred = df_val.copy().reset_index(drop=True)
df_val_pred["pred_label"] = y_pred
df_val_pred["correct"] = df_val_pred["label"] == df_val_pred["pred_label"]

# Show some misclassified hateful memes (label=1, predicted=0)
print("Examples where model FAILED to detect hate (label=1, pred=0):\n")
df_missed_hateful = df_val_pred[(df_val_pred["label"] == 1) & (df_val_pred["pred_label"] == 0)].head(5)
for i, row in df_missed_hateful.iterrows():
    print(f"- TEXT: {row['text']}")
    print(f"  TRUE: Hateful (1), PRED: Non-hateful (0)\n")

# Show some non-hateful memes wrongly flagged (label=0, pred=1)
print("\nExamples where model OVER-flagged (label=0, pred=1):\n")
df_false_positives = df_val_pred[(df_val_pred["label"] == 0) & (df_val_pred["pred_label"] == 1)].head(5)
for i, row in df_false_positives.iterrows():
    print(f"- TEXT: {row['text']}")
    print(f"  TRUE: Non-hateful (0), PRED: Hateful (1)\n")
```

```python
save_dir = "./models/checkpoints/bert_text_baseline"
trainer.save_model(save_dir)
tokenizer.save_pretrained(save_dir)

print("Model and tokenizer saved to:", save_dir)
```
